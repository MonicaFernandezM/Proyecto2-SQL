```python
#Eliminamos columnas de df_spotify que no vamos a utilizar
df_spotify = df_spotify.drop(columns = ['track_id', 'track_number','track_duration_ms', 'explicit',
            'artist_popularity','artist_followers', 'album_id', 'album_total_tracks', 'album_type'])
```

```python
#pasando todo a minusculas para unificar nombres de columnas
df_rock.columns = df_rock.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
#modificamos nombre de la tabla spotify para que los nombres sean iguales
df_spotify = df_spotify.rename(columns= {'track_popularity': 'popularity', 'album_release_date' : 'year', 'album_name': 'album',
                    'track_name': 'track',   'artist_genres': 'genres'})
```

```python
#Cambiamos la columna year en spotify para quedanos solo con el año y lo pasamos a tipo int 
df_spotify['year'] = pd.to_datetime(df_spotify['year'], errors='coerce').dt.year
df_spotify['year'] = df_spotify['year'].astype('Int64')
```

```python
#creando una columna genero para la tabla rock
df_rock['genres'] = 'rock'
df_rock
```

```python
#df_rock_final = pd.concat([df_rock, df_spotify], axis=0, ignore_index=True, join='outer')
```

```python
df_billboard["artist_clean"] = (
    df_billboard["artist_name"]
    .str.lower()
    .str.replace(r" featuring.*", "", regex=True)
    .str.replace(r" feat.*", "", regex=True)
    .str.replace(r"&.*", "", regex=True)
    .str.strip()
)

df_spotify["artist_clean"] = (
    df_spotify["artist_name"]
    .str.lower()
    .str.replace(r" featuring.*", "", regex=True)
    .str.replace(r" feat.*", "", regex=True)
    .str.replace(r"&.*", "", regex=True)
    .str.strip()
)
```

```python
billboard_spotify = df_billboard.merge(
    df_spotify[["artist_clean", "popularity", "genres", "year"]],
    on="artist_clean",
    how="inner"
)

billboard_spotify.head()
```

```python
billboard_spotify.tail()
```

```python
billboard_spotify.shape
```

```python
billboard_spotify_clean = billboard_spotify[[
    "artist_name",
    "artist_clean",
    "year_x",
    "rank",
    "popularity",
    "genres"
]].rename(columns={
    "year_x": "billboard_year"
})
```

```python
billboard_spotify_clean.groupby("artist_name")["popularity"].mean().sort_values(ascending=False).head(10)
```

```python
billboard_spotify_clean.groupby("artist_name", as_index=False)["popularity"].mean()
```

```python
billboard_spotify_clean["decade"] = (billboard_spotify_clean["billboard_year"] // 10) * 10
```

```python
df_rock["artist_clean"] = (
    df_rock["artist"]
    .str.lower()
    .str.strip()
)
```

```python
df_rock_artists = df_rock[["artist_clean"]].drop_duplicates()
```

```python
billboard_spotify_labeled = billboard_spotify_clean.merge(
    df_rock_artists,
    on="artist_clean",
    how="left",
    indicator=True
)
```

```python
billboard_spotify_labeled["is_rock"] = (
    billboard_spotify_labeled["_merge"] == "both"
)

billboard_spotify_labeled.drop(columns="_merge", inplace=True)
```
