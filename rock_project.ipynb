```python
#Elimininamos columnas que no vamos a utilizar
df_rock = df_rock.drop(columns = ['Duration', 'Time_Signature',
       'Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo'])
```

```python
#verificacion de que se eliminaran 
df_rock
```

```python
df_spotify.columns
```

```python
#Eliminamos columnas de df_spotify que no vamos a utilizar
df_spotify = df_spotify.drop(columns = ['track_id', 'track_number','track_duration_ms', 'explicit',
            'artist_popularity','artist_followers', 'album_id', 'album_total_tracks', 'album_type'])
```

```python
#Verificamos que se hayan eliminado
df_spotify
```

```python
#pasando todo a minusculas para unificar nombres de columnas
df_rock.columns = df_rock.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
df_spotify
```

```python
#modificamos nombre de la tabla spotify para que los nombres sean iguales
df_spotify = df_spotify.rename(columns= {'track_popularity': 'popularity', 'album_release_date' : 'year', 'album_name': 'album',
                    'track_name': 'track',   'artist_genres': 'genres'})
```

```python
#verificamos los cambios
df_spotify
```

```python
df_rock
```

```python
#Cambiamos la columna year en spotify para quedanos solo con el año y lo pasamos a tipo int 
df_spotify['year'] = pd.to_datetime(df_spotify['year'], errors='coerce').dt.year
df_spotify['year'] = df_spotify['year'].astype('Int64')
```
