## Limpieza y Unificación de datos 

### Unificando nombre de las columnas y Eliminación 

#### Eliminación columnas - Tabla Rock

```python
#Viendo los nombres de columnas de la tabla rock
df_rock.columns
```

```python
#Elimininamos columnas que no vamos a utilizar en rock
df_rock = df_rock.drop(columns = ['Duration', 'Time_Signature',
       'Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo'])
```

#### Eliminación columnas - Tabla Spotify

```python
#Viendo los nombres de columnas de la tabla spotify
df_spotify.columns
```

```python
#Eliminamos columnas de df_spotify que no vamos a utilizar
df_spotify = df_spotify.drop(columns = ['track_id', 'track_number','track_duration_ms', 'explicit', 
                                        'album_total_tracks', 'album_type'])
```

#### Unificación de columnas 

```python
#pasando todo a minusculas para unificar nombres de columnas rock y billboard
df_rock.columns = df_rock.columns.str.lower().str.strip().str.replace(" ", "_")
df_billboard.columns = df_billboard.columns.str.lower().str.strip().str.replace(" ", "_")
```

#### Modificación de nombres de columnas 

```python
#modificamos nombre de la tabla spotify 
df_spotify = df_spotify.rename(columns= {'track_popularity': 'popularity', 'album_release_date' : 'year', 
                    'artist_genres' : 'genre_name'})
```

```python
#modificamos nombre de la tabla rock
df_rock = df_rock.rename(columns= {'track': 'track_name', 'artist' : 'artist_name', 
                    'album' : 'album_name'})
```

```python
#modificamos nombre de la tabla billboard 
df_billboard = df_billboard.rename(columns={'title': 'track_name',
                                            'artist(s)': 'artist_name',
                                            'no.': 'rank'})
```

### Creación de la columna genre_name en tabla rock, y modificación de tipos de datos

```python
#Cambiamos la columna year en spotify para quedanos solo con el año y lo pasamos a tipo int 
df_spotify['year'] = pd.to_datetime(df_spotify['year'], errors='coerce').dt.year
df_spotify['year'] = df_spotify['year'].astype('Int64')
```

```python
#creando una columna genero para la tabla rock
df_rock['genre_name'] = 'rock'
df_rock
```

## Limpienza de las columnas de artist_name y Verificación de nulos 

### Limpiando nombres de canciones

```python
#Creamos esta funcion para verificar nombres de canciones
def clean_artist_name(s):
    return (
        s.str.lower()
         .str.replace(r" featuring.*", "", regex=True)
         .str.replace(r" feat.*", "", regex=True)
         .str.replace(r"&.*", "", regex=True)
         .str.strip()
    )
```

```python
#Utilizamos la funcion en las tres tablas
df_rock["artist_name"] = clean_artist_name(df_rock["artist_name"])
df_billboard["artist_name"] = clean_artist_name(df_billboard["artist_name"])
df_spotify["artist_name"] = clean_artist_name(df_spotify["artist_name"])
```

### Eliminación de valores nulos en Spotify

```python
#Eliminando filas de year y artist_name con datos faltantes en la tabla spotify
df_spotify = df_spotify.dropna(subset=['year', 'artist_name'])
```

### Verificación de intervalos en la columna year

```python
#los años esten correctos en todas las tablas
df_rock = df_rock[df_rock['year'].between(1950, 2025)]
df_spotify = df_spotify[df_spotify['year'].between(1950, 2025)]
df_billboard = df_billboard[df_billboard['year'].between(1950, 2025)]
```

# Creación del Merge 

## Primer Merge Entre Billboard y Spotify

```python
#Primero realizamos el merge entre billboard y spotify con columnas especificas 
billboard_spotify = df_billboard.merge(
    df_spotify[["artist_name", "popularity", "genre_name", "year"]],
    on="artist_name",
    how="inner"
)

billboard_spotify.head()
```

### Observación de los datos obtenidos 

```python
#Renombramos columnas year de cada tabla
billboard_spotify = billboard_spotify.rename(columns={
    "year_x": "billboard_year",
    "year_y": "spotify_year"
})
```

```python
#Analizamos los datos agrupando 
billboard_spotify.groupby("artist_name")["popularity"].mean().sort_values(ascending=False).head(2)
```

```python
#Analizamos popularidad 
billboard_spotify.groupby("artist_name", as_index=False)["popularity"].mean().head(2)
```

## Merge Final con la tabla Rock

```python
df_merge_final = billboard_spotify.merge(
    df_rock,
    on="artist_name",
    how="left",
    indicator=True
)
```

```python
#observación del merge final 
df_merge_final.head()
```

### ELiminación de duplicados creados al hacerse el merge

```python
df_merge_final =df_merge_final.drop_duplicates()
```

### Observación final de los datos

```python
# Visualizamos los datos obtenidos despues de hacer el merge
du.show_total_info(df_merge_final, 1)
```

```python
# Observación de duplicados
du.mis_dupl(df_merge_final)
```
